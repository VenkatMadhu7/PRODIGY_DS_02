```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'survived': np.random.randint(0, 2, size=100),
    'pclass': np.random.randint(1, 4, size=100),
    'sex': np.random.choice(['male', 'female'], size=100),
    'age': np.random.uniform(1, 80, size=100),
    'sibsp': np.random.randint(0, 6, size=100),
    'parch': np.random.randint(0, 6, size=100),
    'fare': np.random.uniform(10, 500, size=100),
    'embarked': np.random.choice(['C', 'Q', 'S'], size=100)
        }
```

```python
titanic_df = pd.DataFrame(data)
```

```python
titanic_df.loc[5:10, 'age'] = np.nan
titanic_df.loc[15:20, 'embarked'] = np.nan
```

```python
titanic_df['age'].fillna(titanic_df['age'].median(), inplace=True)
```

```python
titanic_df['embarked'].fillna(titanic_df['embarked'].mode()[0], inplace=True)
```

```python
print(titanic_df.isnull().sum())
```

```python
print(titanic_df.describe())
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['fare'], kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='survived', data=titanic_df)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y='survived', data=titanic_df)
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='embarked', y='survived', data=titanic_df)
plt.title('Survival Rate by Embarked')
plt.show()
```
